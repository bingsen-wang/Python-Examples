#Python Code
The Python code illustrate the curve with different periods in x and y directions..

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation,rc,transforms
rc('animation', html='jshtml')

#prameters
Nf = 500 #number of animation frames
t = np.linspace(0,100,Nf*100)
x=np.cos(2*np.pi*np.sqrt(2)*t)
y=np.cos(2*np.pi*np.sqrt(3)*t)

fig = plt.figure(figsize=(9,9))
ax = plt.axes(xlim=(-1.5, 1.5), ylim=(-1.5, 1.5))
fig.tight_layout()
ax.axis('off')
xyplot=ax.plot([],[],lw=1,color='b')[0]

def animate(i): # animation function.  This is called sequentially
  xyplot.set_data(x[:i*100],y[:i*100])
  # xyplot.set_data(x,y)
  return

anim = animation.FuncAnimation(fig, animate, frames=Nf, interval=50)
# anim #uncomment to generate animation in the output area
# to save the animation, uncomment the following three lines
fn = r"xyplot.mp4"
writervideo = animation.FFMpegWriter(fps=20) #save the video at 50 frames per sec
anim.save(fn, writer=writervideo,dpi = 200)
```
