```
from matplotlib import pyplot,animation,cm,patches,transforms;import numpy as np

Nf=120; r1,r2=.3,.7; xy=r2*np.exp(1j*np.linspace(0,2*np.pi,Nf))
cl = list(cm.tab10.colors); fig = pyplot.figure(figsize=([9,9]))
ax = fig.add_axes([0,0,1,1],xlim=[-1.2,1.2],ylim=[-1.2,1.25]); ax.axis('off')
ax.text(0,1.1,"Reorder the Circles",size=40,ha='center')
circles=[patches.Circle((0,0),r1,linewidth=3,fc='none',ec=cl[k%len(cl)],
                        zorder=-k)for k in range(Nf)]
[ax.add_patch(circles[k])for k in range(Nf)]

def animate(i):
  if i%100==0: print(i)
  circles[i//5].set_transform(transforms.Affine2D().translate(r2/5*(i%5+1),0)+ax.transData)
  [circles[k].set_transform(transforms.Affine2D().translate(xy[i//5-k].real,
                              xy[i//5-k].imag)+ax.transData) for k in range(i//5)]
  return

anim = animation.FuncAnimation(fig, animate, frames=Nf*5, interval=20)
anim.save("circles.mp4",writer=animation.FFMpegWriter(fps=60),dpi=200)
```
