#Python Code
The Python code illustrates how to use ArtistAnimation.

```
from numpy import linspace
from matplotlib import animation,pyplot,cm
clst = list(cm.tab20.colors) #colors of lines
t,x = linspace(0,1,500),linspace(1,30,50)
fig = pyplot.figure(figsize=([9,9]))
def animate(i):
  pyplot.plot(t,t**x[i],lw=5-4.5*i/len(x),c=clst[i%len(clst)])
  pyplot.plot(t,t**(1/x[i]),lw=5-4.5*i/len(x),c=clst[i%len(clst)])
  return
anim = animation.FuncAnimation(fig, animate, frames=len(x), interval=50)
anim.save("lines.mp4", writer=animation.FFMpegWriter(fps=10), dpi = 200)
```
