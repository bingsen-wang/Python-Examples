#Python Code
The Python code illustrates how to use ArtistAnimation.

```
from numpy import linspace,logspace
from matplotlib import animation,pyplot,cm,transforms
clst = list(cm.tab20.colors) #colors of lines
# t,x = linspace(0,1,500),linspace(1,30,100)
t,x = linspace(0,1,500),logspace(0,2,100)
fig,ax = pyplot.subplots(figsize=([9,9]))
ax.set_axis_off()
def animate(i):
  rot = transforms.Affine2D().rotate_deg_around(.5,.5,(i+i/len(x))*360/len(x))+ax.transData
  ax.plot(t,t**x[i],lw=3,c=clst[i%len(clst)],transform=rot)
  ax.plot(t,t**(1/x[i]),lw=3,c=clst[i%len(clst)],transform=rot)
  return
anim = animation.FuncAnimation(fig, animate, frames=len(x), interval=50)
anim.save("lines_rotated.mp4", writer=animation.FFMpegWriter(fps=20), dpi = 200)
```
